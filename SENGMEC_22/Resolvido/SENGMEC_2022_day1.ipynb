## Controle de fluxo - if e else

O if  uma estrutura de condio que permite avaliar uma expresso e, de acordo com seu resultado, executar uma determinada ao. No entanto, nenhum comportamento especfico foi definido para o caso de a condio no ser satisfeita. Quando isso  necessrio, precisamos utilizar a reservada else.  extremamente importante se atentar e identao no precisa informa onde comea e onde termina como em matlab ou pseudo-cdigo.

* Sintaxe da estrutura if/else:

```python
if condio:
    #Bloco verdadeiro
else:
    #Bloco falso
```
alm disso, pode-se ter mais de um condio usando `elif`

```python
if condio_1:
    #Bloco 1
elif condio_2:
    #Bloco 2
elif condio_3:
    #Bloco 3
```

````python
#Verificar se o nmero  menor ou maior que zero
a, b = 6, 7

if (a-b)<0:
    print('A subtrao  negativa')
else:
    print('A subtrao  positiva')
````

## Estruturas iterativas - while

S que copiar cdigo sempre	 uma m prtica, queremos escrever o cdigo apenas uma vez. Se queremos repetir o cdigo, fazemos um lao,	ou	um loop, que deve repetir a	instruo dentro de	bloco enquanto ela for verdadeira. O while   esse comando que, assim como o if, recebe	uma	condio. A	diferena 	que	o if, caso a condio seja verdadeira,	executa	apenas uma	vez	o	cdigo de seu	bloco,	j	o while	executa	enquanto a condio	for	verdadeira.

* Sintaxe da estruturas while:

```python
tentativas = 5
while tentativas >= 0:
    #O que deve fazer
    tentativas = tentativas - 1
```
ou de modo equivalente:

```python
tentativas = 5
while True:
    if tentativas >= 0:
        break
    tentativas = tentativas - 1
```

````python
#Elevar todas os valores ao quadrado
lista = [1, 3, 7, 10, 13, 14, 17, 29]

it = 0
while it < len(lista):
    print(lista[it]**2)
    it = it + 1
````

````python
#Encontrar a posio do nmero  13
it = 0
while True:
    if lista[it] == 13:
        break
    it = it + 1
    
print(f'O nmero 13 est na posio {a}')
````

## Exerccio 1 - Sistema escolar

Construa um programa que receba o RA do estudante e busque no banco de dados da escola (dicionrio informado abaixo) seu nome, nota da P1 e P2. Aps, calcule a mdia simples (M) e informe a situao do aluno. Desse modo,

'O aluno **Fulano** de RA:**00000001** possui mdia **9.99** sua situao  **aprovado**' 

Critrio de aprovao dado por:
* Se M>=5  Aprovado 
* Se M<5  Reprovado 

````python
#Dados dos alunos RA(chave) associado ao [nome, P1, P2](valores)
dados = {182154862:['Arthur Chabole', 5, 8],
         171154874:['Gabriel Oliveira', 3, 5],
         212054875:['Luisa Brando', 8, 4],
         221154885:['George lima', 1.5, 3]}
````

````python
RA = 171154874
nome, P1, P2 = dados[RA][0], dados[RA][1], dados[RA][2]

M = (P1 + P2)/2
if M>= 5:
    sit = 'aprovado'
else:
    sit = 'reprovado'
    
print(f'O aluno {nome} de RA:{RA} possui mdia {M} sua situao  {sit}')
````

````python
#Desafio1: faa o pint para todos os RA ... 

lista_RA = [182154862, 171154874, 212054875, 221154885]
for it in lista_RA:
    RA = it
    nome, P1, P2 = dados[RA][0], dados[RA][1], dados[RA][2]

    M = (P1 + P2)/2

    if M>= 5:
        sit = 'aprovado'
    else:
        sit = 'reprovado'

    print(f'O aluno {nome} de RA:{RA} possui mdia {M} sua situao  {sit}')
````

````python
#Desafio2: crie uma funo para receber o RA e mostrar o print()
def sistema_notas(RA):
    nome, P1, P2 = dados[RA][0], dados[RA][1], dados[RA][2]

    M = (P1 + P2)/2

    if M>= 5:
        sit = 'aprovado'
    else:
        sit = 'reprovado'

    print(f'O aluno {nome} de RA:{RA} possui mdia {M} sua situao  {sit}')
    
sistema_notas(182154862)
````

# Funes

Na programao, **funes so blocos de cdigo que realizam determinadas tarefas** que normalmente precisam ser executadas diversas vezes dentro de uma aplicao. Quando surge essa necessidade, para que vrias **instrues no precisem ser repetidas**, elas so agrupadas em uma funo,  qual  dado um nome e que poder ser chamada/executada em diferentes partes do programa.  extremamente importante se atentar e identao.

* Sintaxe de uma funo:

```python
def nome_funo(argumentos):
    #O que a funo deve fazer
    return #O que a funo deve retornar
```
**Cuidado** as variveis criadas dentro de um funo no so vistas pelo cdigo principal somente so vistas aquelas que foram retornas.

````python
#Criando uma funo que soma dois numros
def sub(v1, v2):
    
    resultado = v1 - v2     #O que a funo faz 
    
    return resultado        #O que ela retorna
````

````python
#Utilizando a funo - colocando o nome dos argumentos
sub(v1=3, v2=7)
````

````python
#Utilizando a funo - Por ordem dos argumetos
sub(3, 7)
````

````python
#Utilizando a funo -  Por ordem dos argumetos
sub(7, 3)
````

````python
#Salvando em uma varivel
A = sub(7, 3)
A
````

````python
#Retornando subtrao e a soma
def calculadora(v1, v2):
    
    sub = v1 - v2     #O que a funo faz 
    add = v1 + v2
    
    return sub, add        #O que ela retorna
````

```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
from sympy.abc import *
```

###  Problema 1)  EDO do crescimento exponencial 

$$ \frac{\mathrm{d} y(t)}{\mathrm{d} t} = k*y(t) $$ 

$$ \frac{\mathrm{d} y(t)}{\mathrm{d} t} = k*(1 - \frac{y(t)}{L}) $$

Solução analítica

```python
# Modelando a EDO crescimento irrestrito
y = Function('y')
y = Eq(Derivative(y(t), t), k*y(t))
y
```

```python
# Modelando a EDO crescimento restrito
y_2 = Function('y')
y_2 = Eq(Derivative(y_2(t), t), k*( 1 - y_2(t)/L))
y_2
```

```python
#Solução analítica EDO crescimento irrestrito
sol = dsolve(y)
sol
```

```python
#Solução analítica EDO crescimento irrestrito
sol_2 = dsolve(y_2)
sol_2
```

Solução numérica

```python
# Função cresc irrestrito que retorna dy/dt
def modelo_1(y, t, k):
    dydt = k*y
    return dydt

#Condição inicial
y0 = 1
k = 0.3

#Pontos de tempo
t = np.linspace(0, 10)

#Resolver EDO
y = odeint(modelo_1, y0, t, args=(k,))

#Plots
plt.plot(t,y, label='Modelo crescimento irrestrito')
plt.xlabel('tempo (s)')
plt.ylabel('y(t)')
plt.legend()
plt.grid(linestyle='dotted')
plt.show()
```

```python
# Função cresc irrestrito que retorna dy/dt
def modelo_1(y, t, k):
    dydt = k*y
    return dydt

#Condição inicial
y0 = 1
k = np.linspace(0.1, 0.3, 3)

#Pontos de tempo
t = np.linspace(0, 10)

#Resolver EDO
y1 = odeint(modelo_1, y0, t, args=(k[0],))
y2 = odeint(modelo_1, y0, t, args=(k[1],))
y3 = odeint(modelo_1, y0, t, args=(k[2],))

#Plots
plt.plot(t,y1, label=f'k={k[0]}')
plt.plot(t,y2, label=f'k={k[1]}')
plt.plot(t,y3, label=f'k={k[2]}')
plt.xlabel('tempo (s)')
plt.ylabel('y(t)')
plt.legend()
plt.grid(linestyle='dotted')
plt.show()
```

```python
# Função cresc irrestrito que retorna dy/dt
def modelo_1(y, t, k):
    dydt = k*y
    return dydt

# Função cresc restrito que retorna dy/dt
def modelo_2(y, t, k):
    if t < 5:
        L = 10
    else:
        L = 2
    dydt = k*(1 - y/L)
    return dydt

#Condição inicial
y0 = 1
k = 0.3
k2 = 5

#Pontos de tempo
t = np.linspace(0, 10)

#Resolver EDO
y = odeint(modelo_1, y0, t, args=(k,))
y2 = odeint(modelo_2, y0, t, args=(k2,))

#Plots
plt.plot(t,y, label='Modelo crescimento irrestrito')
plt.plot(t,y2, label='Modelo crescimento restrito')
plt.xlabel('tempo (s)')
plt.ylabel('y(t)')
plt.legend()
plt.grid(linestyle='dotted')
plt.show()
```

###  Problema 2)

$$ \frac{\mathrm{d} y(t)}{\mathrm{d} t} = -y(t) + u(t) $$
$$ y_{0} = 1$$



```python
#Definir o modelo
def modelo_3(y, t):
    if t<10:
        u = 0
    else:
        u = 2
    dydt = (-y + u)/5
    return dydt

#Condições iniciais (pvi)
y0 = 1

#Definindo tempo
t = np.linspace(0, 30)

#Resolvendo a EDO
y = odeint(modelo_3, y0, t)

#Plots
plt.plot(t,y, label='Modelo 3')
plt.xlabel('tempo (s)')
plt.ylabel('y(t)')
plt.legend()
plt.grid(linestyle='dotted')
plt.show()
```

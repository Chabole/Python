# O que so strings?
Strings so **conjuntos ordenados de caracteres**, usados para armazenar e interpretar **informaes textuais**.
Rigorosamente falando, as strings em Python so classificadas como sequncias imutveis, significando que que elas tem uma ordem posicional que vai da esquerda para a direita e no podem ser alteradas no local.
Para escrever uma string normalmente utilizamos **aspas simples ou duplas**, dessa forma o computador est guardando na memria cada um dos caracteres e a sua ordem dentro das aspas simples ou duplas.

Exemplo:

```python
#Sintaxe da string
title = "Semana da engenharia"
title
```

O comando **len()** quando aplicado em uma string, contabiliza o comprimento dela, ou seja, **quantos caracteres ela possui** contando os espaos.

```python
len(title) 
```

### Index

Entendendo o funcionamento das **posies (index) de cada caractere dentro de uma string**. Todas as estruturas de dados em python seguem a mesma lgica e sintaxe, tais como: listas e arrays. 



Dentro da frase ainda podemos **selecionar caracteres especifcos** utilizando *lista[index]*. 

```python
#String 'Luther College'
Faculdade = "Luther College"

#Selecionando por index
a = Faculdade[0]
b = Faculdade[8]
c = Faculdade[10]

print (a,b,c)
```

### Fatiamento
O **fatiamento** significa **selecionar uma parcela da string** dado um comeo e um fim *(string[comeo:fim])*.  O primeiro numero representa o limite inferior e o segundo numero representa o limite superior. Ou seja, se realizarmos um fatiamento [7:13], ser contado de 7 a 12. Exemplo: Faculdade[7:13].

```python
#Fatiamento
Faculdade[7:13], Faculdade[0:6] 
```

### Index com listas
A sintaxe para **acessar um elemento de uma lista**  a mesma usada para acessar um caractere de um string. Ns usamos o operador de indexao (**[ ]** no confundir com a lista vazia). A expresso dentro dos colchetes especifica o ndice. Lembrar que o **ndice do primeiro elemento  0**. Qualquer expresso que tenha como resultado um nmero inteiro pode ser usada como ndice e como com strings, ndices negativos indicaro elementos da direita para a esquerda ao invs de da esquerda para a direita.

*Resumo dos comandos:*
1. **lista[3]** retorna o elementos cuja posio  3
2. **lista[3:9]** retorna os elementos de comeando da posio 3 at a posio 8! (no retorna a posio 9)
3. **lista[3:]** retorna os elementos comeando da posio 3 at o fim da lista
4. **lista[:5]** retorna os elementos comeando da posio 0 at a posio 4. (no retorna a posio 5)

```python
#Lista
x = [3, 5, 6, 88, 9, 12, 90, 2, 0, 1, 5]

#Elemento
x[2]
```

```python
#Fatiamento 
x[:3]
```

Veracidade o comando *in* retorna um **resultado booleano** caso o elementos estiver ou no dentro da lista ou dicionrio.

```python
#Lista de mercado
frutas = ["maca", "laranja", "banana", "cereja"]

#'Elemto' in lista_pesquisar
print("maca" in frutas)
print("pera" in frutas)
```

# O que so dicionrios?

Dicionrio  um tipo **diferente de coleo**. Ele  um tipo de **mapeamento nativo do Python**. Um mapa  uma coleo associativa desordenada. A associao, ou mapeamento,  feita a **partir de uma chave**, que pode ser qualquer tipo imutvel, para um **valor**, que pode ser qualquer objeto de dados do Python no importa em que ordem escrevemos os pares. Os valores em um dicionrio so acessados com chaves, no com ndices, por isso **no h necessidade de se preocupar com a ordenao**.

*Resumo dos comandos:*
1. **{}** - Construindo um dicinrio vazio
1. **{chave_1:valor_1, chave_2:valor_2, chave_3:valor_3}** - sintaxe para construir um dicionrio com dados

```python
#Sequncia chave - valor
dic = {'bala': 3, 'suco': 10}
```

```python
#A partir da chave 'bala' resgatamos o preo
dic['bala']
```

Como exemplo, vamos criar um dicionrio para traduzir palavras em ingls para portugs. Para este dicionrio, as chaves so strings em ingls e os valores em portugus.

```python
#Criando um dicionrio
dic = {'one': 'um', 'two': 'dois', 'three': 'trs', 'four': 'quatro'}
dic
```

```python
#Encontrando valor pela chave
dic['one'], dic['four']
```

```python
#Verificando o que esta contido ou no
print('ones' in dic)
print('five' in dic)
```

```python
#Mostrando os as chaves 
print(dic.keys())

#Mostrando s os valores
print(dic.values())
```

# Funes

Na programao, **funes so blocos de cdigo que realizam determinadas tarefas** que normalmente precisam ser executadas diversas vezes dentro de uma aplicao. Quando surge essa necessidade, para que vrias **instrues no precisem ser repetidas**, elas so agrupadas em uma funo,  qual  dado um nome e que poder ser chamada/executada em diferentes partes do programa. 

```python
#Criando uma funo que soma dois numros
def soma(valor_1, valor_2): #Argumentos da funo
    
    #O que a funo faz 
    soma = valor_1 + valor_2
    
    return soma #O que ela retorna
```

```python
#Utilizando a funo
soma(3, 7)
```

```python
#Criando uma funo que retorna dois numros
def calculadora(valor_1, valor_2): #Argumentos da funo
    
    #O que a funo faz 
    soma = valor_1 + valor_2
    mult = valor_1 * valor_2
    
    return soma, mult #O que ela retorna
```

```python
#Utilizando a funo
calculadora(3,7)
```

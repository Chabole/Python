# O que são strings?
Strings são **conjuntos ordenados de caracteres**, usados para armazenar e interpretar **informações textuais**.
Rigorosamente falando, as strings em Python são classificadas como sequências imutáveis, significando que que elas tem uma ordem posicional que vai da esquerda para a direita e não podem ser alteradas no local.
Para escrever uma string normalmente utilizamos **aspas simples ou duplas**, dessa forma o computador está guardando na memória cada um dos caracteres e a sua ordem dentro das aspas simples ou duplas.

Exemplo:

```python
#Sintaxe da string
title = "Semana da engenharia"
title
```

O comando **len()** quando aplicado em uma string, contabiliza o comprimento dela, ou seja, **quantos caracteres ela possui** contando os espaços.

```python
len(title) 
```

### Index

Entendendo o funcionamento das **posições (index) de cada caractere dentro de uma string**. Todas as estruturas de dados em python seguem a mesma lógica e sintaxe, tais como: listas e arrays. 



Dentro da frase ainda podemos **selecionar caracteres especifícos** utilizando *lista[index]*. 

```python
#String 'Luther College'
Faculdade = "Luther College"

#Selecionando por index
a = Faculdade[0]
b = Faculdade[8]
c = Faculdade[10]

print (a,b,c)
```

```python
#Selecionando por index negativo
Faculdade[-1]
```

### Fatiamento
O **fatiamento** significa **selecionar uma parcela da string** dado um começo e um fim *(string[começo:fim])*.  O primeiro numero representa o limite inferior e o segundo numero representa o limite superior. Ou seja, se realizarmos um fatiamento [7:13], será contado de 7 a 12. Exemplo: Faculdade[7:13].

```python
#Fatiamento
Faculdade[7:13], Faculdade[0:6] 
```

### Métodos built-in 

```python
#Contar quntas letras existem na string
d = Faculdade.count("e")
print(d)

#Colocar tudo maíusculo
e = Faculdade.upper()
print(e)
```

### Index com listas
A sintaxe para **acessar um elemento de uma lista** é a mesma usada para acessar um caractere de um string. Nós usamos o operador de indexação (**[ ]** não confundir com a lista vazia). A expressão dentro dos colchetes especifica o índice. Lembrar que o **índice do primeiro elemento é 0**. Qualquer expressão que tenha como resultado um número inteiro pode ser usada como índice e como com strings, índices negativos indicarão elementos da direita para a esquerda ao invés de da esquerda para a direita.

*Resumo dos comandos:*
1. **lista[3]** retorna o elementos cuja posição é 3
2. **lista[3:9]** retorna os elementos de começando da posição 3 até a posição 8! (não retorna a posição 9)
3. **lista[3:]** retorna os elementos começando da posição 3 até o fim da lista
4. **lista[:5]** retorna os elementos começando da posição 0 até a posição 4. (não retorna a posição 5)

```python
#Lista
x = [3, 5, 6, 88, 9, 12, 90, 2, 0, 1, 5]

#Elemento
x[2]
```

```python
#Fatiamento 
x[:3]
```

Veracidade o comando *in* retorna um **resultado booleano** caso o elementos estiver ou não dentro da lista ou dicionário.

```python
#Lista de mercado
frutas = ["maca", "laranja", "banana", "cereja"]

#'Elemto' in lista_pesquisar
print("maca" in frutas)
print("pera" in frutas)
```

# O que são dicionários?

Dicionário é um tipo **diferente de coleção**. Ele é um tipo de **mapeamento nativo do Python**. Um mapa é uma coleção associativa desordenada. A associação, ou mapeamento, é feita a **partir de uma chave**, que pode ser qualquer tipo imutável, para um **valor**, que pode ser qualquer objeto de dados do Python não importa em que ordem escrevemos os pares. Os valores em um dicionário são acessados com chaves, não com índices, por isso **não há necessidade de se preocupar com a ordenação**.

*Resumo dos comandos:*
1. **{}** - Construindo um dicinário vazio
1. **{chave_1:valor_1, chave_2:valor_2, chave_3:valor_3}** - sintaxe para construir um dicionário com dados

```python
#Sequência chave - valor
dic = {'bala': 3, 'suco': 10}
```

```python
#A partir da chave 'bala' resgatamos o preço
dic['bala']
```

Como exemplo, vamos criar um dicionário para traduzir palavras em inglês para portugês. Para este dicionário, as chaves são strings em inglês e os valores em português.

```python
#Criando um dicionário
dic = {'one': 'um', 'two': 'dois', 'three': 'três', 'four': 'quatro'}
dic
```

```python
#Encontrando valor pela chave
dic['one'], dic['four']
```

```python
#Verificando o que esta contido ou não
print('ones' in dic)
print('five' in dic)
```

```python
#Mostrando os as chaves 
print(dic.keys())

#Mostrando só os valores
print(dic.values())
```

# Funções

Na programação, **funções são blocos de código que realizam determinadas tarefas** que normalmente precisam ser executadas diversas vezes dentro de uma aplicação. Quando surge essa necessidade, para que várias **instruções não precisem ser repetidas**, elas são agrupadas em uma função, à qual é dado um nome e que poderá ser chamada/executada em diferentes partes do programa. 

```python
#Criando uma função que soma dois numéros
def soma(valor_1, valor_2): #Argumentos da função
    
    #O que a função faz 
    soma = valor_1 + valor_2
    
    return soma #O que ela retorna
```

```python
#Utilizando a função
soma(3, 7)
```

```python
#Criando uma função que retorna dois numéros
def calculadora(valor_1, valor_2): #Argumentos da função
    
    #O que a função faz 
    soma = valor_1 + valor_2
    mult = valor_1 * valor_2
    
    return soma, mult #O que ela retorna
```

```python
#Utilizando a função
calculadora(3,7)
```

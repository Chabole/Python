##  Estruturas bsicas de dados - listas
No Python, podemos utilizar diversos tipos de estruturas de dados. Estas estruturas resolvem um tipo de problema e podem ser teis em diversas situaes. As principais estruturas so as Listas, Sets, Dicionrios e Tuplas. Neste curso veremos listas e dicinrios. As listas seguem o mesmo princpio de index que as string mais detalhes em: https://docs.python.org/pt-br/3/tutorial/datastructures.html#more-on-lists

* Sintaxe da lista:

```python
lista = [1, 3, 7, 10, 13, 14, 17, 29]
#ou
lista_lista = [1, 3, ['Zebra','SENGMEC'], [2,3,5] ]
```

*Resumo dos comandos:*
1. **lista[3]** retorna o elementos cuja posio  3
2. **lista[3:9]** retorna os elementos de comeando da posio 3 at a posio 8! (no retorna a posio 9)
3. **lista[3:]** retorna os elementos comeando da posio 3 at o fim da lista
4. **lista[:5]** retorna os elementos comeando da posio 0 at a posio 4. (no retorna a posio 5)
5. **lista[1:9:2]** retorna os elementos comeando da posio 0 at a posio 8 com passo de 2. (no retorna a posio 5)

## Controle de fluxo - if e else

O if  uma estrutura de condio que permite avaliar uma expresso e, de acordo com seu resultado, executar uma determinada ao. No entanto, nenhum comportamento especfico foi definido para o caso de a condio no ser satisfeita. Quando isso  necessrio, precisamos utilizar a reservada else.  extremamente importante se atentar e identao no precisa informa onde comea e onde termina como em matlab ou pseudo-cdigo.

* Sintaxe da estrutura if/else:

```python
if condio:
    #Bloco verdadeiro
else:
    #Bloco falso
```
alm disso, pode-se ter mais de um condio usando `elif`

```python
if condio_1:
    #Bloco 1
elif condio_2:
    #Bloco 2
elif condio_3:
    #Bloco 3
```

````python
#Verificar se o nmero  menor ou maior que zero
a, b = 6, 7

if (a-b)<0:
    print('A subtrao  negativa')
else:
    print('A subtrao  positiva')
````

Critrio de aprovao dado por:

# Funes

Na programao, **funes so blocos de cdigo que realizam determinadas tarefas** que normalmente precisam ser executadas diversas vezes dentro de uma aplicao. Quando surge essa necessidade, para que vrias **instrues no precisem ser repetidas**, elas so agrupadas em uma funo,  qual  dado um nome e que poder ser chamada/executada em diferentes partes do programa.  extremamente importante se atentar e identao.

* Sintaxe de uma funo:

```python
def nome_funo(argumentos):
    #O que a funo deve fazer
    return #O que a funo deve retornar
```
**Cuidado** as variveis criadas dentro de um funo no so vistas pelo cdigo principal somente so vistas aquelas que foram retornas.

````python
#Criando uma funo que soma dois numros
def sub(valor_1, valor_2):
    
    resultado = valor_1 - valor_2     #O que a funo faz 
    
    return resultado                  #O que ela retorna
````

````python
#Utilizando a funo - colocando o nome dos argumentos
sub(valor_1=3, valor_2=7)
````

````python
#Utilizando a funo - Por ordem dos argumetos
sub(3, 7)
````

````python
#Utilizando a funo -  Por ordem dos argumetos
sub(7, 3)
````

````python
#Salvando em uma varivel
A = sub(7, 3)
A
````

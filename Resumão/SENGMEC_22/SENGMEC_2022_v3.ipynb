##  Estruturas básicas de dados - listas
No Python, podemos utilizar diversos tipos de estruturas de dados. Estas estruturas resolvem um tipo de problema e podem ser úteis em diversas situações. As principais estruturas são as Listas, Sets, Dicionários e Tuplas. Neste curso veremos listas e dicinários. As listas seguem o mesmo princípio de index que as string mais detalhes em: https://docs.python.org/pt-br/3/tutorial/datastructures.html#more-on-lists

* Sintaxe da lista:

```python
lista = [1, 3, 7, 10, 13, 14, 17, 29]
#ou
lista_lista = [1, 3, ['Zebra','SENGMEC'], [2,3,5] ]
```

*Resumo dos comandos:*
1. **lista[3]** retorna o elementos cuja posição é 3
2. **lista[3:9]** retorna os elementos de começando da posição 3 até a posição 8! (não retorna a posição 9)
3. **lista[3:]** retorna os elementos começando da posição 3 até o fim da lista
4. **lista[:5]** retorna os elementos começando da posição 0 até a posição 4. (não retorna a posição 5)
5. **lista[1:9:2]** retorna os elementos começando da posição 0 até a posição 8 com passo de 2. (não retorna a posição 5)



````python
#Criando uma lista - [1, 3, 7, 10, 13, 14, 17, 29]
lista = [1, 3, 7, 10, 13, 14, 17, 29]
lista
````

## Controle de fluxo - if e else

O if é uma estrutura de condição que permite avaliar uma expressão e, de acordo com seu resultado, executar uma determinada ação. No entanto, nenhum comportamento específico foi definido para o caso de a condição não ser satisfeita. Quando isso é necessário, precisamos utilizar a reservada else. É extremamente importante se atentar e identação não precisa informa onde começa e onde termina como em matlab ou pseudo-código.

* Sintaxe da estrutura if/else:

```python
if condição:
    #Bloco verdadeiro
else:
    #Bloco falso
```
além disso, pode-se ter mais de um condição usando `elif`

```python
if condição_1:
    #Bloco 1
elif condição_2:
    #Bloco 2
elif condição_3:
    #Bloco 3
```

````python
#Verificar se o número é menor ou maior que zero
a, b = 6, 7

if (a-b)<0:
    print('A subtração é negativa')
else:
    print('A subtração é positiva')
````

## Estruturas iterativas - for 

Existe	 um	loop	 que	 simplifica	 essa	ideia	 de	 começar	 com	 um	 valor e
incrementá-lo	até	chegar	em	um	valor	final:	o	loop for.
Para	entender	o	loop,	ou	laço	for.
A	ideia	é	definirmos	o	valor	inicial	e	o	valor	final,	que	o	loop	o	incrementa	automaticamente

* Sintaxe da estrutura for:

```python
for elemento in lista_elementos:
    #O que deve fazer
```
ou de modo equivalente:

```python
for elemento in range(len(lista_elementos)):
    #O que deve fazer
```



````python
#Elevar todas os valores ao quadrado
lista = [1, 3, 7, 10, 13, 14, 17, 29]

y = []
iteração = 0
while iteração < len(lista):
    y.append(lista[iteração]**2)
    iteração = iteração + 1
y
````

````python
#Encontrar a posição do número  13
a = 0
while True:
    if lista[a] == 13:
        break
    a = a + 1
    
print(f'O número 13 está na posição {a}'
````

````python
#Encontrar a posição do número  13
a = 0
while lista[a] != 13:
    a = a + 1    
print(f'O número 13 está na posição {a}')
````

## Estruturas iterativas - while

Só que copiar código sempre	é uma má prática, queremos escrever o código apenas uma vez. Se queremos repetir o código, fazemos um laço,	ou	um loop, que deve repetir a	instrução dentro de	bloco enquanto ela for verdadeira. O while  é esse comando que, assim como o if, recebe	uma	condição. A	diferença é	que	o if, caso a condição seja verdadeira,	executa	apenas uma	vez	o	código de seu	bloco,	já	o while	executa	enquanto a condição	for	verdadeira.

* Sintaxe da estruturas while:

```python
tentativas = 5
while tentativas >= 0:
    #O que deve fazer
    tentativas = tentativas - 1
```
ou de modo equivalente:

```python
tentativas = 5
while True:
    if tentativas >= 0:
        break
    tentativas = tentativas - 1
```

## O que pode dar errado 

Nem	 sempre	 as	 coisas	 acontecem	 como	 esperado.	 O	 Python	 tem	 uma	 sintaxe	 própria,	 um
vocabulário	 próprio.	 Digitar	 algo	 que	 o	 interpretador	 não	 entende	 causará	 um	 erro	 no	 programa. Não	se	assuste	com	a	mensagem de erro! O python irá dizer onde o erro aconteceu e qual o tipo de erro os principais são **ZeroDivisionError**(divisão por zero), **SyntaxError**(o python não entendeu o que você escreveu muito comum em **for**, **while** e **if**), **IndentationError**(não existe a posição que você esta buscando). Esquecer de fechar parenteses e aspas bem como : em laços de repetição e controle são erros bem comuns.


1. Mais detalhes em: https://docs.python.org/3/library/exceptions.html

# Funções

Na programação, **funções são blocos de código que realizam determinadas tarefas** que normalmente precisam ser executadas diversas vezes dentro de uma aplicação. Quando surge essa necessidade, para que várias **instruções não precisem ser repetidas**, elas são agrupadas em uma função, à qual é dado um nome e que poderá ser chamada/executada em diferentes partes do programa. É extremamente importante se atentar e identação.

* Sintaxe de uma função:

```python
def nome_função(argumentos):
    #O que a função deve fazer
    return #O que a função deve retornar
```
**Cuidado** as variáveis criadas dentro de um função não são vistas pelo código principal somente são vistas aquelas que foram retornas.

````python
#Criando uma função que soma dois numéros
def sub(valor_1, valor_2):
    
    resultado = valor_1 - valor_2     #O que a função faz 
    
    return resultado                  #O que ela retorna
````

````python
#Utilizando a função - colocando o nome dos argumentos
sub(valor_1=3, valor_2=7)
````

````python
#Utilizando a função - Por ordem dos argumetos
sub(3, 7)
````

````python
#Utilizando a função -  Por ordem dos argumetos
sub(7, 3)
````

````python
#Salvando em uma variável
A = sub(7, 3)
A
````

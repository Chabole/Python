## Sympy


**SymPy** é uma biblioteca Python para **matemática simbólica**. Ele tem como objetivo se tornar um **sistema de álgebra computacional (CAS) completo**, mantendo o código o mais simples possível para ser compreensível e facilmente extensível. SymPy é escrito inteiramente em Python.

*Resumo dos comandos:*
1. **Eq(A, B)** - Faz a igualdade entre as expressões A e B 
2. **limit(A, x, $x_0$)** - Encontra o limite da expressão A com respeito a variável x tentendo a um valor $x_0$
3. **A.simplify()** - Simplifica a expressão A
4. **A.evalf()** - Faz uma aproximação numérica da expressão A
5. **solve(B, x)** - Resolve uma expressão B com respeito a variável x

## Escrevendo e resolvendo funções 

Movimento é do tipo MRUV (movimento retilíneo uniformemente variado). Exemplo da queda livre sem resistência do ar.

\begin{equation}
    h(t) = v_{0}.t - \frac{1}{2}g.t^2
\end{equation}

*Resumo dos comandos:*
1. **symbols()** - Constrói simbulos para as expressões  
2. **dif(A, x)** - Encontra a derivada da expressão A em relação a x
3. **integrate(A, x)** - Encontra a integral da expressão A em relação a x
4. **A.subs({D})** - Substitui valores numéricos da expressão A dado um dicionário D

```python
#Definindo simbolos especiais
v0 = symbols('v_0')
```

```python
#Escrevendo uma função
y = v0*t - Rational(1, 2)*g*t**2
y
```

```python
# Achando raizes
raizes = solve(y, t)
raizes
```

### Derivando a função

Derivada de primeira ordem  

\begin{equation}
    \frac{d}{dt} \left ( v_{0}.t - \frac{1}{2}g.t^2 \right )
\end{equation}

```python
# 1ª derivada
dydt = diff(y, t)
dydt
```


Derivada de segunda ordem  

\begin{equation}
    \frac{d}{dt}\left [ \frac{d}{dt} \left ( v_{0}.t - \frac{1}{2}g.t^2  \right )  \right ]
\end{equation}

```python
# 2ª derivada
ddydt = diff(y, t, t)
ddydt
```


Integral indefinidade

\begin{equation}
     \int  \left (v_{0}.t - \frac{1}{2}g.t^2  \right ) dt
\end{equation}

```python
#1ª Integração indefinida
dydt_int = integrate(dydt, t)
dydt_int
```

Integral definidade
\begin{equation}
    \int_{0}^{2} \left ( v_{0}.t - \frac{1}{2}g.t^2 \right ) dt
\end{equation}

```python
#1ª Integração definida
dydt_intd = integrate(dydt, (t, 0, 2))
dydt_intd
```

Integral dupla indefinida

\begin{equation}
    \int \left [ \int  \left (v_{0}.t - \frac{1}{2}g.t^2  \right ) dt \right ]dt
\end{equation}

```python
#Integração dupla
dydt_int = integrate(y, t, t)
dydt_int
```

### Substituindo valores numéricos na função

\begin{equation}
    h(t) = v_{0}.t - \frac{1}{2}g.t^2
\end{equation}


```python
#Substituindo valores numéricos
y.subs({g:9.81, t:0.5, v0:5})
```

```python
#Vetorizando para substituir valores
h = lambdify([t, v0, g], y)
```

```python
#Avaliando numéricamente
h(2, 5, 9.81)
```

```python
#Convertendo para latex 
latex(dydt, mode='equation') # resolver //
```

### Construindo um gráfico

```python
#Criando array de tempo
tempo = np.linspace(0, 5)

#Criando o gráfico
fig, ax = plt.subplots()
ax.set(title='Gráfico da função queda-livre', xlabel='Tempo (s)', ylabel='Altura (m)')

#Fazendo a plotagem da função
ax.plot(tempo, h(tempo, 12, 9.81), label=r'H_1(t)')
ax.plot(tempo, h(tempo, 5, 9.81), label=r'H_2(t)')

#Mostrando a legenda
ax.legend()
#Colocando o grid
ax.grid(linestyle='dotted')
```

## EDO's e classificação
#### Crescimentos irrestrito
Sendo $y(t)$ uma função e $k$ uma constante.

\begin{equation}
    \frac{d}{d t} y{\left(t \right)} = k y{\left(t \right)}
\end{equation}

*Resumo dos comandos:*
1. **Function()** - Constrói uma função genérica desconhecida 
2. **Derivative(y, t)** - Encontra a derivada da expressão y em relação a t para funções genéricas
3. **dsolve(edo)** - Encontra a solução genérica de uma EDO
5. **A.rhs** - Separa a solução dos métodos **dsolve()** e **solve()** para avaliação numérica

```python
#Escrevendo função
y = Function('y')
edo = Eq(Derivative(y(t), t), k*y(t))
edo
```

```python
classify_ode(edo)
```

```python
#solução da edo
sol = dsolve(edo)
sol
```

```python
#Defindo simbolo
C1 = symbols('C1')

#Substituindo valores na solução
sol.subs({C1:0.6, k:2, t:2})
```

```python
#Usando o método rhs
sol.rhs
```

## Sistema massa-mola
Equação diferencial ordinária linear de segunda ordem com coeficientes constantes.

Sendo $y(t)$ a função posição em relação ao tempo $t$, $k$ é a constante da mola  e $c$ constante de amortecimento.

\begin{equation}
   m \frac{d^{2}}{d t^{2}} y{\left(t \right)} + c \frac{d}{d t} y{\left(t \right)} + k y{\left(t \right)}  = 0
\end{equation}

Sendo $m=10kg$ e $k=90$

Avalie para:
1. $c = 100kg/sec$
2. $c = 60kg/sec$
3. $c = 10kg/sec$

```python
#Escrevendo a EDO
y = Function('y')
EDO = Eq(m*Derivative(y(t), t, t) + c*Derivative(y(t), t) + k*y(t), 0) 
EDO
```

```python
#Classificando a EDO
classify_ode(EDO)
```

1. Caso I: $c^2 > 4mk$ com $c = 100kg/sec$
2. Caso II: $c^2 = 4mk$ com $c = 60kg/sec$
3. Caso III: $c^2 < 4mk$ com $c = 10kg/sec$


### Caso I - Superamortecido

```python
#Caso I
EDO_1 = EDO.subs({m:10, k:90, c:100})
EDO_1
```

```python
Sol_1 = dsolve(EDO_1, y(t))
Sol_1.simplify()
```

### Caso II - Amortecimento crítico

```python
#Caso II
EDO_2 = EDO.subs({m:10, k:90, c:60})
EDO_2
```

```python
Sol_2 = dsolve(EDO_2, y(t))
Sol_2.simplify()
```

### Caso III - Sub amortecimento

```python
#Caso III
EDO_3 = EDO.subs({m:10, k:90, c:10})
EDO_3
```

```python
Sol_3 = dsolve(EDO_3, y(t))
Sol_3.simplify()
```

```python
latex(Sol_3.simplify())
```

### Plot das soluções

```python
#Soluções formatadas
C1, C2 = symbols('C1 C2')

#Preparando as soluções para solução numérica
S1 = (Sol_1.rhs).subs({C1:0.5, C2:0.25})
S2 = (Sol_2.rhs).subs({C1:0.5, C2:0.25})
S3 = (Sol_3.rhs).subs({C1:0.5, C2:0.25})

#Vetorizando
S1 = lambdify([t], S1)
S2 = lambdify([t], S2)
S3 = lambdify([t], S3)
```

```python
# Plotando gráficos com as soluções vetorizada
tempo = np.linspace(0, 10, 200)

#Criando o gráfico
fig, ax = plt.subplots()
ax.set(xlabel='Tempo', ylabel='Espaço')

#Fazendo a plotagem da função
ax.plot(tempo, S1(tempo), label='Super amortecido')
ax.plot(tempo, S2(tempo), label='Amortecimento crítico')
ax.plot(tempo, S3(tempo), label='Sub amortecido')

#Mostrando a legenda
ax.legend()
#Colocando o grid
ax.grid(linestyle='dotted')
```

### Manipulando as EDOs e vendo as soluções

```python
#EDO's
EDO_1 = EDO.subs({m:10, k:90, c:100})
EDO_2 = EDO.subs({m:10, k:90, c:60})
EDO_3 = EDO.subs({m:10, k:200, c:10})

#Soluções
Sol_1 = dsolve(EDO_1, y(t))
Sol_2 = dsolve(EDO_2, y(t))
Sol_3 = dsolve(EDO_3, y(t))

#Soluções formatadas
C1, C2 = symbols('C1 C2')

S1 = (Sol_1.rhs).subs({C1:0.5, C2:0.25})
S2 = (Sol_2.rhs).subs({C1:0.5, C2:0.25})
S3 = (Sol_3.rhs).subs({C1:0.5, C2:0.25})

#Vetorizando
S1 = lambdify([t], S1)
S2 = lambdify([t], S2)
S3 = lambdify([t], S3)

# Plotando gráficos com as soluções vetorizada
tempo = np.linspace(0, 5, 200)

fig, ax = plt.subplots()
ax.set(xlabel='Tempo', ylabel='Espaço')

ax.plot(tempo, S1(tempo), label='Super amortecido')
ax.plot(tempo, S2(tempo), label='Amortecimento crítico')
ax.plot(tempo, S3(tempo), label='Sub amortecido')

#Mostrando a legenda
ax.legend()
#Colocando o grid
ax.grid(linestyle='dotted')
```

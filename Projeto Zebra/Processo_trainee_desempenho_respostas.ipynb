## Decolagem 

Perfil da decolagem. O objetivo é calcular a **distância** de decolagem até a velocidade **$V_{lo}$** o ultimo instânte onde há contato com o solo.

Diagrama de corpo livre do avião durante a decolagem. Onde $V_{R}$ é a velocidade com o vento relativo, $L$ força de sustentação, $D$ força de arrasto ou tração requerida, $T$ tração disponível, $F_{at}$ força de atrito com o solo, $N$ normal (relação W-L) e $W$ peso total da aeronave. 

Efeito solo

\begin{equation}
    \phi = \frac{(16h/b)^2}{1 +(16h/b)^2} 
\end{equation}

Coeficientes aerodinâmicos

\begin{equation}
    C_D = (C_{D0} + \phi. K C_{L}^2 )  
\end{equation}

\begin{equation}
    C_{LLO} = \frac{\pi.e_{0}.AR.\mu}{2 \phi}  
\end{equation}


Velocidades importantes 
\begin{equation}
    V_{estol} = \sqrt{\frac{2.W}{\rho . S. C_{L_{max}}}}
\end{equation}

\begin{equation}
    V_{lo} = 1.2 V_{estol}
\end{equation}

Forças aerodinâmicas
\begin{equation}
    D = \frac{1}{2}. \rho. v^2. S. (C_{D0} + \phi. K C_{L}^2 )  
\end{equation}

\begin{equation}
    L = \frac{1}{2}. \rho. v^2. S. C_{L}  
\end{equation}

Aproximação pra distância de decolagem
\begin{equation}
    S_{lo} = \frac{1.44W^2}{g S \rho C_{Lmax} [ T-[D + \mu(W-L)] ] }_{0.7V_{lo}}
\end{equation}

### Desafio 4

Implemente 3 novos métodos dentro da classe avião. Um para calcular a força de sustentação $L$, outro o arrasto $D$ e para calcular o efeito solo $\phi$.

\begin{equation}
    L = \frac{1}{2}. \rho. v^2. S. C_{L}  
\end{equation}

\begin{equation}
    D = \frac{1}{2}. \rho. v^2. S. (C_{D0} + \phi. K C_{L}^2 )  
\end{equation}

\begin{equation}
    \phi = \frac{(16h/b)^2}{1 +(16h/b)^2} 
\end{equation}

Calcule o efeito solo  $\phi$, força de sustentação $L$ e força de arrasto $D$. Para velocidade de 15$m/s$, $C_L = 1.5$ e $\rho = 1.225kg/m^3$.

*Dados:* $b=2.48, S = 0.90m^2, C_{Lmax} = 1.65, C_{D0}=0.12, h=0.35m$

*Algoritmo para programação*
1. Adicione 2 novos atributos no construtor $C_{D0}$ e $h$.
2. Já será calcula pra vc a constante $K$ no construtor.
3. Implemente: **efeito_Solo(self)** é um método que calcula ($\phi$) e busca a altura $h$ como atributo do avião (self.h).
4. Implemente: **sustentação(self, V, CL, rho)** é um método que calcula ($L$) com 3 argumentos de entrada. Velocidade, Cl qualquer, densidade do ar.
4. Implemente: **arrasto(self, V, CL, rho)** é um método que calcula ($D$) com 3 argumentos de entrada. Velocidade, Cl qualquer, densidade do ar.

```python
#Classe 
class avião:
    def __init__(self, S, b, CLmax, C_D0, h): #Construtor da classe
        self.S = S
        self.b = b          #Atributos da classe
        self.CLmax = CLmax      
        self. C_D0 = C_D0
        self.h = h
        
        self.AR = (self.b**2)/self.S 
        self.K = 1/(3.14*0.75*self.AR)
        
    #Método da classe
    def v_Stol(self, W, rho): # Argumentos do método (W e rho)
        V_stol = (2*W/(rho*self.S*self.CLmax))**0.5
        return V_stol
    
    def efeito_Solo(self):
        zw = 16 * self.h /self.b
        phi = (zw ** 2) / (1 + zw ** 2)
        return phi
    
    def sustentação(self, V, CL, rho):
        L = 0.5 * (V ** 2) * rho * self.S * CL
        return L
    
    def arrasto(self, V, CL, rho):
        D = (0.5* (V ** 2)* rho* self.S* (self.C_D0 + self.efeito_Solo() * self.K * CL ** 2))
        return D
```

```python
#Classe 
class avião:
    def __init__(self, S, b, CLmax, C_D0, h): #Construtor da classe
        self.S = S
        self.b = b          #Atributos da classe
        self.CLmax = CLmax      
        self. C_D0 = C_D0
        self.h = h
        
        self.AR = (self.b**2)/self.S 
        self.K = 1/(3.14*0.75*self.AR)
        
    #Método da classe
    def v_Stol(self, W, rho): # Argumentos do método (W e rho)
        V_stol = (2*W/(rho*self.S*self.CLmax))**0.5
        return V_stol
    
    def efeito_Solo(self):
        zw = 16 * self.h /self.b
        phi = (zw ** 2) / (1 + zw ** 2)
        return phi
    
    def sustentação(self, V, CL, rho):
        L = 0.5 * (V ** 2) * rho * self.S * CL
        return L
    
    def arrasto(self, V, CL, rho):
        D = (0.5* (V ** 2)* rho* self.S* (self.C_D0 + self.efeito_Solo() * self.K * CL ** 2))
        return D
    
    def CLL0_Ideal(self, mi):
        cll0 = np.pi*0.717*self.AR*mi/(2*self.efeito_Solo())
        return cll0
```

### Desafio 6
*(Exemplo 4.10 - Pag 217. Miranda)* Determine o comprimento de pista $ S_{lo}$ necessário para a decolagem com um peso total de $150N$ em uma pista localizada ao nível do mar $\rho$ = 
1,225$kg/m^3$.

\begin{equation}
    S_{lo} = \frac{1.44W^2}{g S \rho C_{Lmax} [ T-[D + \mu(W-L)] ] }_{0.7V_{lo}} 
\end{equation}

*Dados:* $b=2.48, S = 0.90m^2, C_{Lmax} = 1.65, C_{D0}=0.12, h=0.35m, \mu=0.03$

*Algoritmo para programação*
1. Implemente: **distancia_decolagem(self, W, rho)** um método que calcula a distância de decolagem (Equação acima).

*Algoritmo de para implementar o método **distancia_decolagem(self, W, rho)***
1. Considere tração disponível $T=25N$
1. Encontre a velocidade de decolagem $V_{lo} = 1.2V_{estol}$ 
2. Calcule o $C_{LLO}$ para substituir em $L$ e $D$. 
3. Calcule $L$ e $D$ na velocidade igual a $0.7V_{lo}$. 
4. Por fim, calcule $S_{lo}$

```python
#Classe 
class avião:
    def __init__(self, S, b, CLmax, C_D0, h): #Construtor da classe
        self.S = S
        self.b = b          #Atributos da classe
        self.CLmax = CLmax      
        self. C_D0 = C_D0
        self.h = h
        
        self.AR = (self.b**2)/self.S 
        self.K = 1/(3.14*0.75*self.AR)
        
    #Método da classe
    def v_Stol(self, W, rho): # Argumentos do método (W e rho)
        V_stol = (2*W/(rho*self.S*self.CLmax))**0.5
        return V_stol
    
    def efeito_Solo(self):
        zw = 16 * self.h /self.b
        phi = (zw ** 2) / (1 + zw ** 2)
        return phi
    
    def sustentação(self, V, CL, rho):
        L = 0.5 * (V ** 2) * rho * self.S * CL
        return L
    
    def arrasto(self, V, CL, rho):
        D = (0.5* (V ** 2)* rho* self.S* (self.C_D0 + self.efeito_Solo() * self.K * CL ** 2))
        return D
    
    def CLL0_Ideal(self, mi):
        cll0 = np.pi*0.717*self.AR*mi/(2*self.efeito_Solo())
        return cll0
    
    def distancia_decolagem(self, W, rho, mi):
        T = 33.207
        
        vlo = self.v_Stol(W, rho)*1.2
        CLLO = self.CLL0_Ideal(mi)
        
        L = self.sustentação(0.7*vlo, CLLO, rho)
        D = self.arrasto(0.7*vlo, CLLO, rho)
        
        slo = (1.44*(W**2))/(9.81*self.S*rho*self.CLmax*(T - (D + mi*(W - L))))
        return slo
```

```python
avião_1 = avião(0.9, 2.48, 1.65, 0.022, 0.35)
avião_1.distancia_decolagem(150, 1.225, 0.03)
```
